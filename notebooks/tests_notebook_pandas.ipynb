```python
import pandas as pd 
import numpy as np
from infra.spark_utils import build_spark
```

```python
df = pd.read_csv('economias.csv', sep=',')
df
```

```python
df['resumo'] = df['resumo'].str.replace(pat="\r\n", repl="\n", regex=False)
df['resumo'] = df['resumo'].str.replace(pat="\r", repl="\n", regex=False)
df
```

```python
df['resumo'] = df['resumo'].str.split(pat="\n", n=-1, expand=False)
df
```

```python
df = df.explode("resumo", ignore_index=True)
df
```

```python
df[["tipo", "ativo", "qtd", "preco_medio", "preco_atual"]] = df['resumo'].str.split(pat="|", expand=True)
for i in ["tipo", "ativo", "qtd", "preco_medio", "preco_atual"]:
    df['i'] = df['i'].str.strip()
df
```

```python
df["data_apuracao"] = pd.to_datetime(df["data_apuracao"], format="%d/%m/%Y")
df["ano"] = df["data_apuracao"].dt.year
df["mes"] = df["data_apuracao"].dt.month_name(locale="pt_BR.utf8")
df
```

```python
tickers_internacionais = ['BERK34', 'IVVB11', 'AAPL34', 'MSFT34', 'NVDC34', 'GOGL34', 'AMZO34', 'TSLA34', 'META34', 'MELI34']
df['exposicao'] = np.where(df['ativo'].isin(tickers_internacionais), "internacional", "nacional")
df
```
